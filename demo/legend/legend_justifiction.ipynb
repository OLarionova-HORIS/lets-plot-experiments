```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
mpg = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```python
p = ggplot(mpg, aes('displ', 'cty', )) + \
    geom_point(aes(color='drv', size='hwy')) + \
    scale_size(range=[1, 10], breaks=[15, 30, 40]) + \
    theme_bw() + \
    theme(legend_background=element_rect(size=1), 
          plot_background=element_rect(size=1))
p
```

```python
gggrid([
    p + theme(legend_justification = [0,0]),
    p + theme(legend_justification = [0,1]),
    p + theme(legend_justification = [0,0.5])
])
```

```python
gggrid([
    p + theme(legend_position='top', legend_justification = [0,0]),
    p + theme(legend_position='top', legend_justification = [1,0]),
    p + theme(legend_position='top', legend_justification = [0.5,0])
], ncol=1)
```

```python
# inside

gggrid([
    p + theme(legend_position=[0,0], legend_justification = [0,0]),
    p + theme(legend_position=[0,1], legend_justification = [0,1]),
])
```

```python
gggrid([
  p + theme(legend_position = "left", legend_justification = [0,0.5]),
  p + theme(legend_position = [0.0,0.5], legend_justification = [0,0.5])
])
```

#### legend_margin 

```python
# Set 0 for margins

p + theme(legend_justification = [0,1], legend_margin=[0])
```

```python
# Remove space between legend and panel plot

p + theme(legend_justification = [0,1], legend_margin=[0], plot_inset=[6.5,0,6.5,6.5])
```
