```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
DEF_ANGLE = 90.0
DEF_CURVATURE = 0.5
DEF_NCP = 5

def curve_plot(curvature=DEF_CURVATURE, angle=DEF_ANGLE, ncp=DEF_NCP):
    return ggplot() \
            + geom_point(aes('x', 'y'), data = { 'x': [-30, 30], 'y': [-10,10] }) \
            + geom_curve(x = -20, y = 1, xend = 10, yend = -1, 
                         curvature = curvature, angle = angle, ncp = ncp,
                         arrow=arrow(ends='both'), tooltips=layer_tooltips().line("Tooltip")) \
            + ggtitle("curvature={0}, angle={1}, ncp={2}".format(curvature, angle, ncp))
```

```python
gggrid([
    curve_plot(ncp=1),
    curve_plot(ncp=2),
    curve_plot(ncp=5),
    curve_plot(ncp=15)
], ncol=2)
```

```python
gggrid([
    curve_plot(curvature=1, angle=45, ncp=1),
    curve_plot(curvature=1, angle=45, ncp=2),
    curve_plot(curvature=1, angle=45, ncp=5),
    curve_plot(curvature=1, angle=45, ncp=10)    
], ncol=2)
```
