```r
library("ggplot2")
```

```r
options(repr.plot.height=4,repr.plot.width=8)
theme_update(title = element_text(size=18), text = element_text(size=16))
```

```r
df <- data.frame(
  trt = factor(c(1, 1, 2, 2)),
  resp = c(1, 5, 3, 4),
  group = factor(c(1, 2, 1, 2)),
  upper = c(1.1, 5.3, 3.3, 4.2),
  lower = c(0.8, 4.6, 2.4, 3.6)
)
```

```r
ggplot(df, aes(trt, resp, colour = group)) + 
   geom_errorbar(aes(ymin = lower, ymax = upper), linewidth=2, width = 0.2) + 
   ggtitle("geom_errorbar()")
```

```r
ggplot(df, aes(trt, resp, colour = group)) + 
   geom_errorbar(aes(ymin = lower, ymax = upper), linewidth=2, width = 0.2) +
   coord_flip() +
   ggtitle("coord_flip()")
```

```r
ggplot(df, aes(resp, trt, colour = group)) + 
    geom_errorbar(aes(xmin = lower, xmax = upper), width = 0.2, linewidth=2, orientation='y') +
    ggtitle("y-orientation")
```

```r
ggplot(df, aes(resp, trt, colour = group)) +
  geom_errorbar(aes(xmin = lower, xmax = upper), width = 0.2, linewidth=2) +
  ggtitle("Flip the orientation by changing mapping")
```

```r
ggplot(df, aes(resp, trt, colour = group)) + 
    geom_errorbarh(aes(xmin = lower, xmax = upper), height = 0.2, linewidth=2) + 
    ggtitle("geom_errorbarh()")
```

### Use  position_dodge

```r
ggplot(df, aes(trt, resp, colour = group)) +
    geom_errorbar(aes(ymin = lower, ymax = upper), width = 0.2, linewidth=2,
                  position=position_dodge(0.4)) +
   ggtitle("position = dodge")
```

```r
ggplot(df, aes(resp, trt, colour = group)) + 
    geom_errorbar(aes(xmin = lower, xmax = upper), width = 0.2, linewidth=2, orientation='y', 
                  position=position_dodge(0.4)) +
   coord_flip() +
   ggtitle("coord_flip()")
```

```r
ggplot(df, aes(resp, trt, colour = group)) + 
    geom_errorbar(aes(xmin = lower, xmax = upper), width = 0.2, linewidth=2, orientation='y', 
                  position=position_dodge(0.4)) +
    ggtitle("y-orientation")
```

```r
ggplot(df, aes(resp, trt, colour = group)) + 
    geom_errorbarh(aes(xmin = lower, xmax = upper), height = 0.2, linewidth=2,
                   position=position_dodge(0.4)) +
    ggtitle("geom_errorbarh")
```
