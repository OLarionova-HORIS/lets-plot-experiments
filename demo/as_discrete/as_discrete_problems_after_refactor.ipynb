```python
from lets_plot import *
from lets_plot.mapping import as_discrete
```

```python
LetsPlot.setup_html(offline=True, no_js=True)
```

```python
df1 = {
    'x': [0, 5, 10],
    'y': [0, 5, 10],
    'a': [3, 1, 2],
    'b': [5, 6, 4]
}
```

```python
p = ggplot(df1, aes(x='x', y='y'))

gggrid([
    p + geom_point(aes(fill='a', color='b'), size=6, shape=21) + ggtitle("fill='a', color='b'"),
    p + geom_point(aes(fill=as_discrete('a'), color='b'), size=6)+ ggtitle("fill=as_discrete('a'), color='b'"),
    p + geom_point(aes(fill='a', color=as_discrete('b')), size=6)+ ggtitle("fill='a', color=as_discrete('b')"),
    p + geom_point(aes(fill=as_discrete('a'), color=as_discrete('b')), size=6)+ ggtitle("fill=as_discrete('a'), color=as_discrete('b')")
],
    ncol=2
) + ggsize(1000,700)
```

#### Mapping to the same variable

```python
# color and fill to 'a' => the same column with the same status
gggrid([
    p + geom_point(aes(color='a', fill='a'), size=6, shape=21),
    p + geom_point(aes(color='a', fill=as_discrete('a', order=1)), size=6, shape=21)
]) + ggsize(1000,300)
```

### Order by 'as_discrete'-variable

```python
p + geom_point(aes(fill=as_discrete('a'), color=as_discrete('b', order_by='a', order=1)), size=6)
```

### Bar examples

```python
data = {
    "a" : ["a","a","b"],
    "b" : [14,21,32]
}

gggrid([
    ggplot(data, aes(x="a", y="b", fill=as_discrete("b"))) + geom_bar(),
    ggplot(data, aes(x="a", fill=as_discrete("b"))) + geom_bar()
])
```

```python
gggrid([
    ggplot(data, aes(x="a", y="b", fill="b")) + geom_bar(stat='identity'),
    ggplot(data, aes(x="a", y="b", fill=as_discrete("b"))) + geom_bar(stat='identity')
])
```
