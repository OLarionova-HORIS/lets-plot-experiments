```python
from lets_plot import *
import pandas as pd
```

```python
LetsPlot.setup_html()
```

```python
mpg2 = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg2.csv")
mpg2.head()
```

```python
p = ggplot(mpg2, aes('vehicle weight (lbs.)', 'miles per gallon', label='vehicle name'))

p + geom_label()
```

```python
p + geom_label(check_overlap=True)
```

```python
# Rotate

gggrid([
    p + geom_label(angle = 15),
    p + geom_label(angle = 15, check_overlap=True)
])
```
