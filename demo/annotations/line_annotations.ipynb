```python
import pandas as pd
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
iris_df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/iris.csv')
```

```python
ggplot(iris_df) + geom_line(aes('sepal_length', color='species'), 
                            stat='density',
                            labels=layer_labels().line('^color').size(15))
```

```python
ggplot(iris_df) + \
    geom_line(aes('sepal_length', color='species'), 
              stat='density',
              labels=layer_labels().line('^color').size(15)) + \
    xlim(4.5, 6.5) 
```

```python
n = 1000
np.random.seed(42)
x = np.random.normal(size=n)
y = np.random.normal(size=n)

ggplot({'x': x, 'y': y}, aes('x', 'y')) + \
    geom_density2d(aes(color='..group..'), size=1, show_legend=False,
                  labels=layer_labels().line('^color')) + \
    scale_color_gradient(low='red', high='dark_green')
```
