```r
library("ggplot2")
```

```r
# install.packages("ggrepel")
library(ggrepel)
```

```r
options(repr.plot.width = 10, repr.plot.height = 6) 
```

```r
set.seed(42)

dat <- subset(mtcars, wt < 5.0)
dat$car <- rownames(dat)

p <- ggplot(dat, aes(wt, mpg, label = car)) + geom_point(color = "red")

p + geom_text() + labs(title = "geom_text()")
```

```r
p + geom_text_repel() + labs(title = "geom_text_repel()")
```

```r
p + geom_text_repel(angle=60) + labs(title = "geom_text_repel(angle=60)")
```

```r
p + geom_text_repel(max.overlaps = 3)
```

```r
p + geom_text_repel(box.padding = 0.8)
```

```r
p + geom_text_repel(box.padding = 0.8, point.padding = 1.5)
```

```r
p + geom_text_repel(
    box.padding = 1.8, 
    segment.curvature = -0.2,
    arrow = arrow(length = unit(0.015, "npc"))
)
```
