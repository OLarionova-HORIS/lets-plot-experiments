```r
library("ggplot2")
```

```r
options(repr.plot.height=4,repr.plot.width=10)
theme_update(title = element_text(size=18), text = element_text(size=20))
```

```r
df <- data.frame(
  x = c(0, 0, 0),
  a = c('a', 'a', 'b')
)
```

```r
ggplot(df, aes(x, y=x)) + geom_count()
```

```r
# geom_point() with stat='sum'

ggplot(df, aes(x, y=x, color=a), size=after_stat(n)) + geom_point(stat='sum', position=position_dodge(0.1))
```

```r
# geom_count()

ggplot(df, aes(x, y=x, color=a)) + geom_count(position=position_dodge(0.1))
```

```r
ggplot(mpg, aes(cty, hwy)) + geom_point(aes(size=..n..), stat='sum')
```

Default

```r
ggplot(mpg, aes(x=class, y=drv)) + geom_count()
```

Display proportions instead of counts:

```r
ggplot(mpg, aes(x=class, y=drv)) + geom_count(aes(size=..prop..)) 
```

Scale to the proportion within each class:

```r
ggplot(mpg, aes(x=class, y=drv)) + geom_count(aes(size=..prop.., group=class)) 
```

```r
ggplot(mpg, aes(x=class, y=drv)) + 
  geom_count(aes(size=..prop..), colour="lightgrey") +
  geom_count(aes(size=..prop.., group=class), colour="cornflowerblue")  +
  scale_size(range = c(0,10), breaks=seq(0,1,by=0.2)) +
  coord_fixed() +
  theme_minimal()
```
