```r
library("ggplot2")
```

```r
options(repr.plot.height=4,repr.plot.width=10)
theme_update(title = element_text(size=18), text = element_text(size=20))
```

```r
df <- data.frame(
  x = c(0, 0, 0),
  a = c('a', 'a', 'b')
)
```

```r
ggplot(df, aes(x, y=x)) + geom_count()
```

```r
ggplot(df, aes(x, y=x, color=a)) + geom_count(position=position_dodge(0.1))
```

```r
ggplot(df, aes(x, y=x, color=a), size=..n..) + geom_point(stat='sum', position=position_dodge(0.1))
```
