```r
library(ggplot2)
```

```r
theme_set(theme_gray(base_size = 20))
options(repr.plot.height=5,repr.plot.width=8)
```

```r
data <- data.frame(
  x = c(1, 2, 3, 4),
  y = c(10, 5, 8, 12),
  category = c("A", "B", "A", "B")
)

p1 <- ggplot(data, aes(x, y, color = category)) + geom_point(size = 3)
p1
```

```r
p1 + scale_color_discrete(labels = c('A' = 'Alpha', 'B' = 'Beta'))
```

```r
#  ?

p1 + scale_color_discrete(breaks = c('Alpha' = 'A', 'Beta' = 'B'))
```

```r
p2 <- p1 + theme(legend.position='none')

p2 + scale_x_continuous(breaks = c("a" = 2, "b" = 3))
```

```r
breaks <- c(1,2,3,4)
names(breaks) <- c('a','b','c','d')

p2 + scale_x_continuous(labels = names(breaks))
```
